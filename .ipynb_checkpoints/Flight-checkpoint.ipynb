```python
from requests_html import HTMLSession #import requests_html 软件包的HTMLSession功能
```

```python
session = HTMLSession() #建立一个session
```

```python
url = 'https://zh.flightaware.com/live/flight/CSC8633/history/20180513/2205Z/ZUCK/ZUUU/tracklog'
```

```python
r = session.get(url) #get url的网站信息
```

```python
import pandas as pd
import numpy as np
```

```python
#r = session.get("https://zh.flightaware.com/live/flight/CSC8633/history/20180513/2205Z/ZUCK/ZUUU/tracklog")
```

```python
sel = 'tr' #选择网站里的<tr>路径
```

```python
results = r.html.find(sel) #读它找到的sel数据
```

```python
results 
```

```python
def get_text_from_sel(sel):
    mylist = []
    try:
        results = r.html.find(sel)
        for result in results:
            mytext = result.text
            
            mylist.append((mytext))
        return mylist
    except:
        return None
    
#读取sel路径里的数据
```

```python
print(get_text_from_sel(sel)) #打印结果
```
